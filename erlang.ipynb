```python
from scipy.stats import erlang
```

```python
rv=erlang(a=144,scale=600)
```

The length of a block intervals can be modeled using an Erlang distribution (https://en.wikipedia.org/wiki/Erlang_distribution) with shape a and scale 600. 

```python
from datetime import timedelta
```

```python
def quantiletime(q):
    return str(timedelta(seconds=rv.ppf(q)));
```

So for CI, I guess you mean equal quantiles on both ends of the distribution:

```python
def CI(x):
    half=(1-x/100)/2
    return quantiletime(half), quantiletime(1-half)
```
